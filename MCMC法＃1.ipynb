−１〜１までの一様乱数を使って点を発生させる

　　　　　　　　↓
        
 その点が中心が原点で半径１の円の内部にあるか判定する
 
 点を100個発生したときの円の内部にある点は76個だったとする
 
 76/100 = π/4
 π ＝　76/25 = 3.04
 
 点を多く発生させればπ本来の値に近づく

```python
#モンテカルロ法
import numpy as np
import matplotlib.pyplot as plt

#乱数を固定
np.random.seed(1)

#円作成
x,y =[],[]

#円を作成するために360度分割する
for d in np.linspace(-180,180,360):
    x.append(np.sin(np.deg2rad(d)))
    y.append(np.cos(np.deg2rad(d)))
    
#点を発生
N = 1000

#-1から1までN個の乱数を発生する
xmc = np.random.uniform(-1,1,N)
ymc = np.random.uniform(-1,1,N)

#原点からの距離を計算
r = (xmc ** 2 + ymc ** 2)**0.5

#円の中に点を１、円の外にある点を０とする
accept = np.where(r<=1,1,0)

#円の中に点がある確率を求める
accept_ratio = np.sum(accept)/N

print(accept_ratio)
```

```python
print("数値的に計算 π/4:" ,accept_ratio)
print("真の値　π/4:", np.pi/4)
print("数値的に計算　π:",accept_ratio*4)
print("真の値　π:",np.pi)
```

```python
#円の描画
#グラフを正方形にしている
plt.gca().set_aspect("equal",adjustable = "box")
plt.plot(x,y)
#点の描画
plt.scatter(xmc,ymc,color="red",marker=".")
plt.show()
```
